```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_fill_between(arr):
    means = np.zeros(len(arr[0]))
    stds = np.zeros(len(arr[0]))
    for index in range(len(arr[0])):
        means[index] = np.mean([i[index] for i in arr])
        stds[index] = np.std([i[index] for i in arr])
    return means, stds

lstm1 = [0.16367171932388755, 0.20235400166466294, 0.2700911924306507, 0.13143902991972597, 0.19790258524290466]
lstm2 = [0.16690099846868703, 0.20512201508475747, 0.276174504799639, 0.13603054571609788, 0.21961553084783017]

means,stds = get_fill_between([lstm1,lstm2])
plt.plot([str(i) for i in range(1,6)],means, '-',alpha=1)
plt.fill_between([str(i) for i in range(1,6)],means - stds, means + stds,alpha=0.2)
```
